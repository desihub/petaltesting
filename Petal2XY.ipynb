```python
import gspread
import pandas as pd
from gspread_dataframe import get_as_dataframe
from oauth2client.service_account import ServiceAccountCredentials
```

```python
def plot_hole_info(ax,hole_coords,nons,fifs):
    for i in range(len(hole_coords)):
        x = hole_coords[i][0]
        y = hole_coords[i][1]

        if i not in nons:
            ax.plot(x, y, color = 'lightgrey', marker='o', zorder = -1)
            if i in fifs:
                text = 'F' + str(i)
                col = 'blue'
            elif i in gifs:
                text = 'G' + str(i)
                col = 'purple'
            else:
                text = i
                col = 'black'
            ax.text(x-.1, y + 0.3, text, color = col, fontsize=10)
```

```python
PETAL=2
petal_id=str(PETAL).zfill(2)

# nominal hole location data
hole_coords = np.genfromtxt('/Users/parkerf/Research/DESI/PetalTesting/hole_coords.csv', delimiter = ',', usecols = (3,4), skip_header = 40)
nons = {38, 331, 438, 460, 478, 479, 480, 481, 497, 498, 499, 500, 513, 514, 515, 516, 527, 528, 529, 530, 531, 535, 536, 537, 538, 539, 540}
gifs = {541, 542}
fifs = {11, 75, 150, 239, 321, 439, 482, 496, 517, 534}

# Google sheet
url = 'https://docs.google.com/spreadsheets/d/1lJ9GjhUUsK2SIvXeerpGW7664OFKQWAlPqpgxgevvl8/edit#gid=0'
credentials = '/Users/parkerf/Research/DESI/PetalTesting/google_access_account.json'
scope = ['https://spreadsheets.google.com/feeds', 'https://spreadsheets.google.com/feeds', 'https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name(credentials, scope)
client = gspread.authorize(creds)
sheet = client.open_by_url(url).sheet1
df  = get_as_dataframe(sheet, parse_dates=True, usecols=[0,1,2,3,4,6,8,14], skiprows=20, header=0)
pdf=df[df['PETAL_ID'] == int(petal_id)]

dev_list=pdf['DEVICE_ID'].tolist()
hole_list=pdf['DEVICE_LOC'].tolist()
dev_id_loc=dict(zip(dev_list,hole_list))
```

```python
def grade_pos(max_blind,max_corr, rms_corr, max_corr_95, rms_corr_95):
    if (max_blind <= 100) & (max_corr <=15) & (rms_corr <=5):
        grade = 'A'
    elif (max_blind <= 250) & (max_corr <=25) & (max_corr_95 <=15) & (rms_corr <=10) & (rms_corr <=5):
        grade = 'B'
    elif (max_blind <= 250) & (max_corr <=50) & (max_corr_95 <=25) & (rms_corr <=20) & (rms_corr <=10):
        grade = 'C'
    elif (max_blind <= 500) & (max_corr <=50) & (max_corr_95 <=25) & (rms_corr <=20) & (rms_corr <=10):
        grade = 'D'
    else:
        grade = 'F'
        
    return grade
```

```python
Grades = {}
for i, t in enumerate(Tests):
    Corr_RMS = []
    Blind = []
    Pos = []
    BAD = []

    name = tests[i]
    Grades[name] = {}
    for pos, df in DATA.items():
        td = np.abs(t - df.index)
        x = np.argmin(td)
        ddf = df.iloc[[x]]
        rms_corr = float(ddf['corr rms (um) all targets with 5.0 um threshold'])
        max_corr = float(ddf['corr max (um) all targets with 5.0 um threshold'])
        max_blind = float(ddf['blind max (um) all targets'])
        max_corr_95 = float(ddf['corr max (um) best 95% with 5.0 um threshold'])
        rms_corr_95 = float(ddf['corr rms (um) best 95% with 5.0 um threshold'])

        grade = grade_pos(max_blind,max_corr, rms_corr, max_corr_95, rms_corr_95)
        Grades[name][pos] = grade
        
        if grade == 'F':
            BAD.append(int(pos[2:]))
        else:
            Corr_RMS.append(rms_corr)
            Blind.append(max_blind)
            Pos.append(int(pos[2:]))
        
        
    holes=[dev_id_loc[e] for e in Pos]
    x = [hole_coords[int(e)][0] for e in holes]
    y = [hole_coords[int(e)][1] for e in holes]

    holesb=[dev_id_loc[e] for e in BAD]
    xb = [hole_coords[int(e)][0] for e in holesb]
    yb = [hole_coords[int(e)][1] for e in holesb]
                        

    fig, (ax1) = plt.subplots(1, figsize = (20,10))
    plot_hole_info(ax1,hole_coords,nons,fifs)

    sc1=ax1.scatter(x,y,s=120,c=Corr_RMS)
    ax1.scatter(xb,yb,marker='*',s=120, c='red')
    ax1.set_title(name+": Corr RMS")
    fig.colorbar(sc1, ax=ax1)
    
    fig, (ax1) = plt.subplots(1, figsize = (20,10))
    plot_hole_info(ax1,hole_coords,nons,fifs)

    sc1=ax1.scatter(x,y,s=120,c=Blind)
    ax1.scatter(xb,yb,marker='*',s=120, c='red')
    ax1.set_title(name+": Blind Max")
    fig.colorbar(sc1, ax=ax1)
    
    plt.figure()
    plt.hist(Blind)
    plt.title("Blind Move Histogram")
    plt.figure()
    plt.hist(Corr_RMS)
    plt.title("Corr RMS Histogram")
```

```python
FF = {}
for test, gs in Grades.items():
    print(test)
    p = []
    c = []
    d = []
    f = []
    ff = []
    FF[test] = []
    for pos, g in gs.items():
        if (g == 'A') | (g == 'B'):
            p.append(pos)
        elif g == 'C':
            c.append(pos)
        elif g == 'D':
            d.append(pos)
        elif g == 'F':
            f.append(pos)
        if (g == 'F') and (int(pos[2:]) not in disabled):
            ff.append(pos)
            FF[test].append(int(pos[2:]))

            
    print("Grade A/B: %d" % len(p))
    print("Grade C #%d: " %len(c), np.sort(c))
    print("Grade D #%d: "%len(d), np.sort(d))
    print("Grade F #%d: "%len(f), np.sort(f))
    print("Grade F not disabled #%d: "%len(ff), np.sort(ff))
    print('\n\n')
```
